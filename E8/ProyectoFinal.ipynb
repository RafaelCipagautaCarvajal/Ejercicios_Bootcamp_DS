#### 4. Ingeniería de atributos y normalización

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, classification_report, roc_auc_score
```

```python
# Seleccionar variables relevantes
X = df.drop(columns=["default"])
y = df["default"]
print(X.columns)
```

```python
# Normalización
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Aplicar SMOTE para balanceo de clases
from imblearn.over_sampling import SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)
```

```python
print(f"Cantidad de registros después de SMOTE: {X_resampled.shape[0]}")
print(f"Distribución de clases después de SMOTE:\n{pd.Series(y_resampled).value_counts()}")
```

-------------------------------
#### 5. División de datos y entrenamiento de modelos

```python
# Dividir datos en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)
```

```python
# Entrenamiento de modelos
models = {
    "Random Forest": RandomForestClassifier(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(n_estimators=100, random_state=42)
}

best_model = None
best_auc = 0

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    auc = roc_auc_score(y_test, y_proba)
    print(f"{name} - Accuracy: {accuracy_score(y_test, y_pred):.4f}, AUC: {auc:.4f}")
    if auc > best_auc:
        best_auc = auc
        best_model = model
```

-------------------------------
#### 6. Optimización de hiperparámetros

```python
# Optimización de hiperparámetros para el mejor modelo
param_grid = {
    "n_estimators": [50, 100, 200],
    "max_depth": [None, 10, 20]
}

grid_search = GridSearchCV(best_model, param_grid, cv=5, scoring='roc_auc')
grid_search.fit(X_train, y_train)
```

```python
# Mejor modelo final
final_model = grid_search.best_estimator_
```

-------------------------------
#### 7. Evaluación del modelo

```python

y_final_pred = final_model.predict(X_test)
y_final_proba = final_model.predict_proba(X_test)[:, 1]

print("\nMejor modelo después de optimización:")
print(f"Accuracy: {accuracy_score(y_test, y_final_pred):.4f}, AUC: {roc_auc_score(y_test, y_final_proba):.4f}")
print("\nReporte de clasificación:")
print(classification_report(y_test, y_final_pred))
```

-------------------------------
#### 8. Matriz de confusión

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_final_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["No Default", "Default"])
disp.plot(cmap="Blues")
plt.title("Matriz de Confusión")
plt.show()
```

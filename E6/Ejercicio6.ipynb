```python
import requests
import pandas as pd

url = "https://www.datos.gov.co/resource/ptgf-ywrb.json?$limit=605000"
#Ya que por defecto solo se muestran 1000 registros, y el dataset en el portal de Datos Abiertos Colombia indica que existen 603.232 filas se especifica un límite de 605000 registros.
headers = {"Accept-Encoding": "gzip, deflate"}

response = requests.get(url, headers=headers)
print(response)
data = response.json()
```

```python
dataf=pd.DataFrame(data)
dataf
```
